# Load Data:

```python
import numpy as np
import pandas as pd
data =  pd.read_csv('compresive_strength_concrete.csv')
data
```

```python
data.isnull().sum() #no null values
```

```python
data.info()
```

```python
from sklearn import preprocessing
d = preprocessing.normalize(data)
names = ['Cement','Blast Furnace Slag','Fly Ash','Water','Superplasticizer','Coarse Aggregate','Fine Aggregate','Age','ConcretecCompressiveStrength']
scaled_df = pd.DataFrame(d, columns=names)
scaled_df.head()
```

```python
#we are going to check correlation between independent and dependent variables such that the x variables with high correlation with y will be taken further for prediction of y.
import seaborn as sns
import matplotlib.pyplot as plt
corrmatrics = scaled_df.corr()
highcorr_features =  corrmatrics.index
plt.figure(figsize=(20,20))
plot = sns.heatmap(scaled_df[highcorr_features].corr(),annot=True,cmap='RdYlGn')
```

```python
x = scaled_df.loc[:,scaled_df.columns != 'ConcretecCompressiveStrength']
y = scaled_df.loc[:,scaled_df.columns == 'ConcretecCompressiveStrength']

# 30% Test(Samples,Labels):
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state=42)
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
network = models.Sequential()
network.add(layers.Dense(10, activation='relu', input_shape=(8,)))
network.add(layers.Dense(8, activation='relu'))
network.add(layers.Dense(6, activation='relu'))
network.add(layers.Dense(1))
network.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
network.fit(x_train, y_train, epochs=100, batch_size=128)
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
network1 = models.Sequential()
network1.add(layers.Dense(2, activation='relu', input_shape=(8,)))
network1.add(layers.Dropout(0.2))
network1.add(layers.Dense(1))
network1.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
network1.fit(x_train, y_train, epochs=100, batch_size=128)
```

```python
#Evaluation Step:
test_loss, test_acc = network.evaluate(x_test, y_test) #before adding drop out layer
print('test_acc:', test_acc)

test_loss, test_acc = network1.evaluate(x_test, y_test) #after adding drop out layer
print('test_acc:', test_acc)
```

```python
#Prediction:
network.predict(x_test) #prediction is negative as the values are less than 0.5 
```
